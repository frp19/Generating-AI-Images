```python
import random

rooms = ['living room', 'kichen', 'bedroom', 'bathroom', 'laundry', 'dining room']
styles = ['modern', 'rustic', 'industrial', 'minimalist', 'traditional']
color_schemes = ['soft natural tones', 'vibrant colors', 'monochrome', 'pastels', 'dark tones']
furnitures = ['sectional sofa', 'armchair', 'bean bags', 'ottoman', 'chaise lounge']
furniture_positions = ['center', 'right corner', 'left corner']
art_decorations = ['abstract painting', 'modern sculpture', 'vintage prints', 'minimalist artwork', 'colorful tapestry']
lightings = ['modern floor lamp', 'chandelier', 'pendant lights', 'table lamps', 'hidden LED strips']
views = ['cityscape', 'garden view', 'beach front', 'snowy landscape', 'mountain view']


# Function to generate a prompt
def generate_prompt():
    style = random.choice(styles)
    color_scheme = random.choice(color_schemes)
    furniture = random.choice(furnitures)
    art_decoration = random.choice(art_decorations)
    lighting = random.choice(lightings)
    view = random.choice(views)
    furniture_position = random.choice(furniture_positions)

    prompt = (
        f"Imagine a {style}, designed with {color_scheme}. "
        f"The room features a {furniture} in the {furniture_position}, accented with {art_decoration}. "
        f"A {lighting} provides ambient lighting. Large windows offer a {view}, "
        "complemented by the interior decor."
    )
    return prompt


# Generate and print multiple prompts
num_prompts = 10  # Change this to generate more or fewer prompts
prompts = [generate_prompt() for _ in range(num_prompts)]
# for _ in range(num_prompts):
    # print(generate_prompt())
prompts
```

```python
from diffusers import UnCLIPPipeline
from diffusers import StableDiffusionPipeline
from diffusers import AutoPipelineForText2Image

import torch


models = dict(
    # stable= StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16),
    un_cli = UnCLIPPipeline.from_pretrained("kakaobrain/karlo-v1-alpha", torch_dtype=torch.float16),
    # auto = AutoPipelineForText2Image.from_pretrained("kandinsky-community/kandinsky-2-2-decoder", torch_dtype=torch.float16)
    )
models = {key:val.to("cuda") for key, val in models.items()}
```

```python
loop_no = 5
prompt_no = 2
for model_name, model in models.items():
  for p_i, p in enumerate(prompts):
    for l in range(loop_no):
      images = model([p for _ in range(prompt_no)]).images
      for i, image in enumerate(images):
        print(model_name, p_i, i)
        image.save(f"image_generated/{model_name}_prompt_{p_i}_no_{i+(prompt_no*l)}.png")
```
